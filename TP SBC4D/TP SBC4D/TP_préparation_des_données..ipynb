```python
df = pd.read_excel('HDR 2007-2008 Table 26.xlsx', header = 9)
df = df.dropna(how="all", axis=1)
df = df.dropna(how="all", axis=0).reset_index(drop=True)
df.head(6)
```

```python
columns = [
    "HDI rank",
    "Country",
    "Internally displaced peoplea(thousands)2006e ",
    "Reference 1",
    "Refugees By country of asylum(thousands)2006e",
    "Reference 2",
    "Refugees By country of originc(thousands) 2006e",
    "Reference 3",
    "Conventional arms transfersb(1990 prices) Imports (US$ millions) 1996",
    "Reference 4",
    "Conventional arms transfersb(1990 prices) Imports (US$ millions)2006",
    "Reference 5",
    "Conventional arms Exports US$ millions  2006",
    "Reference 6",
    "Conventional arms Exports Shared(%)  2002-06",   
    "Total armed forces Thousands2007", 
    "Reference 7",
    "Total armed forces Index (1985 = 100)2007"
]
df.columns = columns
```

```python
#Enregistrer les informations autres que les données contenues dans le dataframe

df_notes = df.iloc[197:199,:]
df_src = df.iloc[199:200,]

df_notes = df_notes.dropna(how = 'any', axis = 1).reset_index(drop=True)
df_src = df_src.dropna(how = 'any', axis = 1).reset_index(drop=True)

df_notes
```

```python
df_src
```

```python
# Effacer les notes et les sources du tableau principal
df = df.drop(np.arange(180,len(df)), axis=0).reset_index(drop=True)
```

```python
hdi_categories = ["HIGH HUMAN DEVELOPMENT", "MEDIUM HUMAN DEVELOPMENT ", "LOW HUMAN DEVELOPMENT"]
# Se débarrasser des rangées avec les catégories HDI

df = df.drop(df[df["HDI rank"].isin(hdi_categories)].index, axis=0).reset_index(drop=True)
```

```python
# Référence pour np.where https://docs.scipy.org/doc/numpy/reference/generated/numpy.where.html
# Pensez que cela ressemble à une déclaration du type "if" : where(test est vrai, vrai-quelque chose, faux-quelque chose)
# ceci peut aussi être imbriqué : where(test is true, true-thing, where(new-test is true, true-thing, false-thing))
df.loc[:, "HDI category"] = np.where(df["HDI rank"] > 155,
                                     "LOW HUMAN DEVELOPMENT",
                                     np.where(df["HDI rank"] < 70,
                                              "HIGH HUMAN DEVELOPMENT",
                                              "MEDIUM HUMAN DEVELOPMENT"
                                             )
                                    )
df.sort_values(by="HDI rank", inplace=True)
df.head()
```

```python
# Enregistrer les informations dans un nouveau fichier excel 

with pd.ExcelWriter('HDR_2007-2008Table_26_modifie.xlsx') as writer:
    df.to_excel(writer, sheet_name = 'Table_26', index = False)
    df_notes.to_excel(writer, sheet_name = 'NOTES', index = False)
    df_src.to_excel(writer, sheet_name = 'SOURCES', index = False)
```
